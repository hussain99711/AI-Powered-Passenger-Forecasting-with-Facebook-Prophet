```python
import pandas as pd
import seaborn as sns

# Load a clean dataset (flights data)
df = sns.load_dataset('flights')

# Show first 5 rows
df.head()

# Display first few rows
df.head()

# See all column names
print("Columns:", df.columns.tolist())

# Check shape (rows x columns)
print("Shape:", df.shape)

# Check for missing values
print("Missing values:\n", df.isnull().sum())

# Basic statistics
df.describe(include='all')

# Convert month column to string first
df['month'] = df['month'].astype(str)

# Now combine into date
df['date'] = df['month'] + ' ' + df['year'].astype(str)
df['date'] = pd.to_datetime(df['date'], format='%b %Y')

# Set date as index and drop the old columns
df.set_index('date', inplace=True)
df.drop(['year', 'month'], axis=1, inplace=True)

# Preview updated DataFrame
df.head()

!pip install prophet

# Prophet requires specific column names
df_prophet = df.reset_index().rename(columns={'date': 'ds', 'passengers': 'y'})
df_prophet.head()

from prophet import Prophet
import matplotlib.pyplot as plt

# Step 1: Create and train the model
model = Prophet()
model.fit(df_prophet)

# Step 2: Create future dates (e.g. next 12 months)
future = model.make_future_dataframe(periods=12, freq='M')

# Step 3: Predict future values
forecast = model.predict(future)

# Step 4: Plot the forecast
fig1 = model.plot(forecast)
plt.title("Passenger Forecast")
plt.show()

# Prophet requires specific column names
df_prophet = df.reset_index().rename(columns={'date': 'ds', 'passengers': 'y'})
df_prophet.head()

from prophet import Prophet
import matplotlib.pyplot as plt

# Step 1: Create and train the model
model = Prophet()
model.fit(df_prophet)

# Step 2: Create future dates (e.g. next 12 months)
future = model.make_future_dataframe(periods=12, freq='M')

# Step 3: Predict future values
forecast = model.predict(future)

# Step 4: Plot the forecast
fig1 = model.plot(forecast)
plt.title("Passenger Forecast")
plt.show()


```
